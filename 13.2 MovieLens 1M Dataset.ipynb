```python
import numpy as np
import pandas as pd
```

```python
pd.options.display.max_rows = 10
```

```python
unames=['user_id','gender','age','occupation','zip']
users = pd.read_table('datasets/movielens/users.dat', sep='::', header=None, names=unames)
mnames=['movie_id', 'title', 'genres']
movies = pd.read_table('datasets/movielens/movies.dat', sep='::', header=None, names=mnames)
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('datasets/movielens/ratings.dat', sep='::',header=None, names=rnames)
```

```python
ratings
```

```python
data = ratings.merge(users, left_on='user_id', right_on='user_id')
data = data.merge(movies, left_on='movie_id', right_on='movie_id')
data
```

```python
grouped = data[['title','gender','rating']].groupby(['title','gender'])
```

```python
i=0;
for a,b in grouped:
    print(a)
    print(b)
    i+=1;
    if (i==2):
        break
```

```python
def average(group):
    return group['rating'].mean()
```

```python
mean_rating = grouped.apply(average).unstack()
mean_rating
```

```python
title_size = data.groupby('title').size()
above_250 = title_size.index[title_size>=250]
above_250
```

```python
mean_rating = mean_rating.loc[above_250].sort_values(by='F',ascending=False)
```

```python
mean_rating[:10]
```

```python
mean_rating['diff'] = mean_rating['M'] - mean_rating['F'] 
```
